### Estruturas de Dados
Com base nestas listas, foi criado um dicionário em que o valor inicial definido para cada habilidade elencada é 0.
Em sequência, utilizando a biblioteca pandas, este dicionário foi transformado em DataFrame (estrutura de dados tabulares). Vale ressaltar que aquelas carreiras que não apresentavam determinada habilidade, o valor ao invés de ser declarado como 0, fica como **NaN** (valor definido em Python para acusar ausência de valor numérico - *Not a Number*.  

```python
import pandas as pd

# Criando dicionários para cada categoria
categories = {
    'back_end': back_end,
    'front_end': front_end,
    'mobile': mobile,
    'game': game,
    'crm': crm,
    'cloud': cloud,
    'web_IA': web_IA,
    'data_analytics': data_analytics,
    'data_engineering': data_engineering,
    'infra_devops_security': infra_devops_security,
    'qualidade': qualidade,
    'lideranca': lideranca
}

# Convertendo as listas em dicionários com valor 0
dict_representation = {}
for cat, skills in categories.items():
    dict_representation[cat] = {skill: 0 for skill in skills}

# Convertendo em DataFrame
df = pd.DataFrame.from_dict(dict_representation, orient='index')

# Os valores faltantes serão NaN por padrão
print(df)
```

### Preparação, extração, manipulação e limpeza dos dados
Até esta etapa, foi realizada a preparação para a análise a ser feita. Criada a lista com as palavras-chave, posteriormente um dicionário com o valor inicial 0 para cada chave-valor. A preparação foi concluída após transformar esse dicionário em uma estrutura de dados que o pandas manipula de forma exemplar: o DataFrame.

Agora, utilizando a biblioteca nativa do Python *os*, será feita a leitura de cada arquivo .md. Esses arquivos constam a apresentação individual de cada aluno/desenvolvedor que está participando deste Desafios de Projeto **Crie Um Portfólio Vencedor**.

Este é o momento de extração e manipulação dos dados, o *processamento* das informações de interesse. Feita a leitura de cada [README.md](https://pt.wikipedia.org/wiki/Readme) pessoal, o algoritmo compara se os arquivos em questão contém alguma das palavras-chaves definidas para cada carreira. Caso encontre, adiciona o valor 1 em cada respectiva coluna de habilidades.

```python
import os

# Juntar todas as listas em uma única lista
all_keywords = back_end + front_end + mobile + game + crm + cloud + web_IA + data_analytics + data_engineering + infra_devops_security + qualidade + lideranca

# Remover duplicados (se houver)
all_keywords = list(set(all_keywords))

# Lendo e comparando o conteúdo dos arquivos
path_to_files = "../community/"  
for filename in os.listdir(path_to_files):
    with open(os.path.join(path_to_files, filename), 'r') as f:
        file_content = f.read()
        
        # Verificar cada keyword
        for keyword in all_keywords:
            if keyword in file_content:
                # Adiciona 1 para a categoria correspondente (se o valor não é NaN)
                if keyword in back_end and df.loc['back_end', keyword] is not None:
                    df.loc['back_end', keyword] += 1
                if keyword in front_end and df.loc['front_end', keyword] is not None:
                    df.loc['front_end', keyword] += 1
                if keyword in mobile and df.loc['mobile', keyword] is not None:
                    df.loc['mobile', keyword] += 1
                if keyword in game and df.loc['game', keyword] is not None:
                    df.loc['game', keyword] += 1
                if keyword in crm and df.loc['crm', keyword] is not None:
                    df.loc['crm', keyword] += 1
                if keyword in cloud and df.loc['cloud', keyword] is not None:
                    df.loc['cloud', keyword] += 1
                if keyword in web_IA and df.loc['web_IA', keyword] is not None:
                    df.loc['web_IA', keyword] += 1
                if keyword in data_analytics and df.loc['data_analytics', keyword] is not None:
                    df.loc['data_analytics', keyword] += 1
                if keyword in data_engineering and df.loc['data_engineering', keyword] is not None:
                    df.loc['data_engineering', keyword] += 1
                if keyword in infra_devops_security and df.loc['infra_devops_security', keyword] is not None:
                    df.loc['infra_devops_security', keyword] += 1
                if keyword in qualidade and df.loc['qualidade', keyword] is not None:
                    df.loc['qualidade', keyword] += 1
                if keyword in lideranca and df.loc['lideranca', keyword] is not None:
                    df.loc['lideranca', keyword] += 1
```

###  Limpeza dos dados
Os dados preparados, extraídos e manipulados, faremos a limpeza destes dados. Cada DataFrame apresenta em suas linhas, cada registro equivalente a uma *carreira tech*. Suas colunas correspondem aos atributos e tecnologias essenciais a cada carreira. Porém, nem todas carreiras apresentam os mesmos atributos. Desta forma, devemos substituir os valores *NaN* por 0, para que se possa realizar a contabilidade, a soma de qual carreira tech apresentou a maior quantidade de atributos ou habilidades mencionadas nos README.md individual. 

Assim, podemos realizar uma **análise de perfil de carreira**, qual a carreira que mais se encaixa nas descrições de todos os participantes do curso.

```python
# Substituir NaN por 0
df.fillna(0, inplace=True)

# Somar os valores por linha
df['Total'] = df.sum(axis=1)

# Mostrar as três linhas com os maiores valores na coluna 'Total'
top_3 = df.nlargest(3, 'Total')

top_3.head()
```

```python
print('As Carreiras Tech que mais aparecem nos arquivos .md individuais são:')

print()

for index in top_3.index:
    if index == 'back_end':
        print('Back-End')
    elif index == 'front_end':
        print('Front-End')
    elif index == 'mobile':
        print('Mobile')
    elif index == 'game':
        print('Game Development')
    elif index == 'crm':
        print('CRM Systems')
    elif index == 'cloud':
        print('Cloud Services')
    elif index == 'web_IA':
        print('Web 3.0 & AI')
    elif index == 'data_analytics':
        print('Data Analytics')
    elif index == 'data_engineering':
        print('Data Engineering')
    elif index == 'infra_devops_security':
        print('Infrastructure, DevOps, & Security')
    elif index == 'qualidade':
        print('Quality Assurance')
    elif index == 'lideranca':
        print('Leadership')
    else:
        print(index)  # Este é um fallback caso haja algum índice que não mapeamos.
```
